```python
# Find value_counts from which N% of total values is concentrated

def fold_80(df, col=df.columns):
    for col in col:
        threshold = 0.2
        items = np.array(df.groupby(col).size().value_counts().sort_index(ascending=True))
        occs = df[col].value_counts(ascending=True).unique()
        nulls = df[col].isnull().sum()
        rest = ((items * occs).sum() + nulls) * threshold
        acc = np.array([items[i] * occs[i] for i in range(len(occs))]).cumsum()
        ix = sum(1 for x in takewhile(lambda x : x < rest, acc))
        return occs[ix]
```

```python
# Create new column for days_to_booking, check values

df['days_to_booking'] = (pd.to_datetime(df['booking_date']) - pd.to_datetime(df['lead_date'])).dt.days
print(df['days_to_booking'].value_counts().sort_index(ascending=True), '\n''\n',
      'Nulls: ', df['days_to_booking'].isnull().sum())
```

```python
# column to values 0 to 1

df['days_to_booking'] = MinMaxScaler().fit_transform(df[['days_to_booking']]).round(3)

print(df['days_to_booking'].value_counts().sort_index(ascending=True))
df['days_to_booking'].hist(bins=60);
```

```python
# as no values between 0 and 1, and 0 is the best, let's give a positive, minimum value

df.loc[df['days_to_booking'] < df['days_to_booking'].value_counts().index[1], 'days_to_booking'] = \
        df['days_to_booking'].value_counts().sort_index(ascending=True).index[1] / \
        (df['days_to_booking'].max() + df['days_to_booking'].max())


# remove last value 1.0 as it's an outlier

df.loc[df['days_to_booking'] > df['days_to_booking'].value_counts().sort_index(ascending=True).index[-2],
       'days_to_booking'] = np.nan

print(df['days_to_booking'].value_counts().sort_index(ascending=True))
df['days_to_booking'].hist(bins=60);
```

```python
# invert values so the best, the biggest.

df['days_to_booking'] = (1 - df['days_to_booking'])

df['days_to_booking'].value_counts().sort_index(ascending=True)
```

```python
for col in to_reduce:
    fold80 = fold_80(df, col=[col])
    print(col, '\n',
          'Categories to reduce: ',
          df[col].value_counts()[df[col].value_counts() < fold80].count(), '\n',
          'Categories to keep: ',
          df[col].value_counts()[df[col].value_counts() >= fold80].count(), '\n',  
         '% categories to reduce: ',
         ((df[col].value_counts()[df[col].value_counts() < fold80].count()/df[col].nunique())*100)\
          .round(decimals=2), '\n',
          '% of rows: ',
         ((df[col][df[col].map(df[col].value_counts()) < fold80].value_counts().sum()/len(df))*100)\
          .round(decimals=2),
         '\n')
```

# Functions for preprocessing

```python
def preproc_clean_main(df, feature_reduction=False, feature_selection=False):
    '''
    Preprocessing for raw data
    
    - feature_reduction as True includes reduced-cardinality features instead raw ones
    - reduced_dataset includes only top features
    
    '''
    
# Create new_clientid for new and returning converters
    df['new_cid'] = np.where(df['clientid'].duplicated(keep=False), 'returning', 'new')
    
    
# creates weekday from date  
    df['weekday'] = pd.to_datetime(df['lead_date']).dt.strftime("%A")
    

# cleans counters
    df['phone_count'] = np.where(df['lead_count'] > df['phone_count'], df['lead_count'], df['phone_count'])
    df['postcode_count'] = np.where(df['phone_count'] > df['postcode_count'], df['phone_count'], df['postcode_count'])
    df['starts_count'] = np.where(df['postcode_count'] > df['starts_count'], df['postcode_count'], df['starts_count'])
    df['entries_count'] = np.where(df['entries_count'] == 0, 1, df['entries_count'])
    df['web_events'] = df['entries_count'] + df['starts_count'] + df['postcode_count'] + df['phone_count'] + df['lead_count']
    df['web_events'] = np.where(df['web_events'] == 5, 'single', 'multiple')


# concatenates steps    
    df.loc[:, 'steps_all'] = df['step0'] + '-' + df['step1'] + '-' + df['step2'] + '-' + df['step3'] + '-' +\
                                df['step4'] + '-' + df['step5'] + '-' + df['step6'] + '-' + df['step7'] + '-' +\
                                df['step8'] + '-' + df['step9'] + '-' + df['step10']

    
# cleans campaigns
    df['camp'] = np.where(df['campaign_cd'].isnull(), df['campaign'], df['campaign_cd'])
    

# creates days_to_booking from dates and camp

    #create new column by lead & booking dates difference
    df['days_to_booking'] = (pd.to_datetime(df['booking_date']) - pd.to_datetime(df['lead_date'])).dt.days

    # column to values 0 to 1
    df['days_to_booking'] = ((df['days_to_booking'] - df['days_to_booking'].min()) / \
                              (df['days_to_booking'].max() - df['days_to_booking'].min())).round(2).astype(float)    

    # as no values between 0 and 1, and 0 is the best, let's give a positive, minimum value
    df.loc[df['days_to_booking'] < df['days_to_booking'].value_counts().index[1], 'days_to_booking'] = \
            df['days_to_booking'].value_counts().sort_index(ascending=True).index[1] / \
            (df['days_to_booking'].max() + df['days_to_booking'].max())


    # remove last value 1.0 as it's an outlier
    df.loc[df['days_to_booking'] > df['days_to_booking'].value_counts().sort_index(ascending=True).index[-2],
           'days_to_booking'] = np.nan
    
    # explode days_to_booking to camp level by target mean replacement
    df['days_to_booking'] = df.groupby('camp')['days_to_booking'].transform(np.mean)


# cleans mediums adding a base 'rest'
    df['clean_medium'] = 'rest'

    mask_0 = (df['camp'].str.lower().str.contains('fb|facebook|instagram', na=False) |
              df['source'].str.lower().str.contains('fb|facebook|instagram', na=False))
    mask_1 = df['camp'].str.lower().str.contains('generic', na=False)
    mask_2 = df['camp'].str.lower().str.contains('marca', na=False)
    mask_3 = (df['camp'].str.lower().str.contains('seo', na=False) |
              df['medium'].str.lower().str.contains('seo|organic', na=False))
    mask_4 = df['camp'].str.lower().str.contains('discv|display|discovery|gdn', na=False)
    mask_5 = df['camp'].str.lower().str.contains('af_', na=False)
    mask_6 = (df['camp'].str.lower().str.contains('mail', na=False) |
              df['medium'].str.lower().str.contains('mail', na=False))

    df.loc[mask_0, 'clean_medium'] = 'facebook'
    df.loc[mask_1, 'clean_medium'] = 'search_generic'
    df.loc[mask_2, 'clean_medium'] = 'search_brand'
    df.loc[mask_3, 'clean_medium'] = 'seo'
    df.loc[mask_4, 'clean_medium'] = 'display'
    df.loc[mask_5, 'clean_medium'] = 'affiliates'
    df.loc[mask_6, 'clean_medium'] = 'email'

    
# creates reduced features

    # camp
    camp_counts = df['camp'].value_counts()
    df['camp_reduced'] = np.where(df['camp'].\
                                  isin(camp_counts.index[camp_counts < fold_80(df, col=['camp'])]), 'reduced', df['camp'])

    # device_browser
    device_browser_counts = df['device_browser'].value_counts()
    df['device_browser_reduced'] = np.where(df['device_browser'].\
                                            isin(device_browser_counts.\
                                                 index[device_browser_counts < fold_80(df, col=['device_browser'])]),
                                            'reduced', df['device_browser'])

    # device_mobile_model
    device_mobile_model_counts = df['device_mobile_model'].value_counts()
    df['device_mobile_model_reduced'] = np.where(df['device_mobile_model'].\
                                                 isin(device_mobile_model_counts.\
                                                      index[device_mobile_model_counts < fold_80(df,
                                                                                                 col=['device_mobile_model'])]),
                                                 'reduced', df['device_mobile_model'])

    # device_mobile_branding
    device_branding_counts = df['device_mobile_branding'].value_counts()
    df['device_mobile_branding_reduced'] = np.where(df['device_mobile_branding'].\
                                                    isin(device_branding_counts.\
                                                         index[device_branding_counts < fold_80(df,
                                                                                                col=['device_mobile_branding'])]),
                                                    'reduced', df['device_mobile_branding'])
    # device_operating_system_version
    device_osv_counts = df['device_operating_system_version'].value_counts()
    df['device_osv_reduced'] = np.where(df['device_operating_system_version'].\
                                        isin(device_osv_counts.\
                                             index[device_osv_counts < fold_80(df,
                                                                               col=['device_operating_system_version'])]),
                                        'reduced', df['device_operating_system_version'])
    
    
# selects relevant columns for each case

    df_clean = df.filter(items=['new_cid', 'weekday', 'source', 'clean_medium', 'camp',
                                'device_category', 'device_browser', 'device_browser_version', 'device_operating_system',
                                'device_operating_system_version', 'device_mobile_branding','device_mobile_model',
                                'step0', 'step1', 'step2', 'step3', 'step4', 'step5', 'step6', 'step7',
                                'step8', 'step9', 'step10', 'steps_all', 'is_booking'])

# check arguments
    
    if feature_reduction:
        df_clean['camp'] = df['camp_reduced']
        df_clean['device_browser'] = df['device_browser_reduced']
        df_clean['device_operating_system_version'] = df['device_osv_reduced']
        df_clean['device_mobile_branding'] = df['device_mobile_branding_reduced']
        df_clean['device_mobile_model'] = df['device_mobile_model_reduced']

        
    elif feature_selection:
        df_clean = df_clean.filter(items=['camp', 'clean_medium', 'device_browser_version',
                                          'device_mobile_model', 'steps_all', 'is_booking'])
        
    return(df_clean)
```
