## Encoding both data sources

```python
# open csv and create X and y
csv_clean = 'data_files/preproc_clean_cookieless.csv'
df_clean = pd.read_csv(csv_clean)
X = df_clean.loc[:, df_clean.columns != 'is_booking']
y = df_clean['is_booking']

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# columns by type
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object', 'bool']).columns
```

### One Hot Encoding

```python
# preproc pipeline with ohe
pipe_ohe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='other')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))])

# fit on train, transform train and test
pipe_ohe.fit(X_train, y_train)
X_train_ohe = pipe_ohe.transform(X_train)
X_test_ohe = pipe_ohe.transform(X_test)
```

```python
#find fair hyperparms for target ecoding on a dummy rfc
pipeline = Pipeline(steps= [('encoding', TargetEncoder()),
                            ('classifier', DecisionTreeClassifier())])
params = {
    'encoding__drop_invariant': [False, True],
    'encoding__min_samples_leaf': [1, 5, 10, 25, 50],
    'encoding__smoothing': [0., 0.01, 0.1]}

gs = GridSearchCV(estimator = pipeline,
                  param_grid = params,
                  cv=3,
                  n_jobs=-1)

gs.fit(X_train, y_train)

#unpack best params and print
te_bp = gs.best_params_
te_bp = {k[10:]: v for (k, v) in te_bp.items()}
print(te_bp)
```

### Target Encoding with Cross Validation

```python
# create encoder from https://brendanhasz.github.io/2019/03/04/target-encoding#cross-fold-target-encoding

class TargetEncoderCV(TargetEncoder):
    """Cross-fold target encoder.
    """
    
    def __init__(self, n_splits=3, shuffle=True, cols=None):
        """Cross-fold target encoding for categorical features.
        
        Parameters
        ----------
        n_splits : int
            Number of cross-fold splits. Default = 3.
        shuffle : bool
            Whether to shuffle the data when splitting into folds.
        cols : list of str
            Columns to target encode.
        """
        self.n_splits = n_splits
        self.shuffle = shuffle
        self.cols = cols
        

    def fit(self, X, y):
        """Fit cross-fold target encoder to X and y
        
        Parameters
        ----------
        X : pandas DataFrame, shape [n_samples, n_columns]
            DataFrame containing columns to encode
        y : pandas Series, shape = [n_samples]
            Target values.
            
        Returns
        -------
        self : encoder
            Returns self.
        """
        self._target_encoder = TargetEncoder(cols=self.cols)
        self._target_encoder.fit(X, y)
        return self

    
    def transform(self, X, y=None):
        """Perform the target encoding transformation.

        Uses cross-fold target encoding for the training fold,
        and uses normal target encoding for the test fold.

        Parameters
        ----------
        X : pandas DataFrame, shape [n_samples, n_columns]
            DataFrame containing columns to encode

        Returns
        -------
        pandas DataFrame
            Input DataFrame with transformed columns
        """

        # Use target encoding from fit() if this is test data
        if y is None:
            return self._target_encoder.transform(X)

        # Compute means for each fold
        self._train_ix = []
        self._test_ix = []
        self._fit_tes = []
        kf = KFold(n_splits=self.n_splits, shuffle=self.shuffle)
        for train_ix, test_ix in kf.split(X):
            self._train_ix.append(train_ix)
            self._test_ix.append(test_ix)
            te = TargetEncoder(cols=self.cols)
            if isinstance(X, pd.DataFrame):
                self._fit_tes.append(te.fit(X.iloc[train_ix,:],
                                            y.iloc[train_ix]))
            elif isinstance(X, np.ndarray):
                self._fit_tes.append(te.fit(X[train_ix,:],
                                            y[train_ix]))
            else:
                raise TypeError('X must be DataFrame or ndarray')

        # Apply means across folds
        Xo = X.copy()
        for ix in range(len(self._test_ix)):
            test_ix = self._test_ix[ix]
            if isinstance(X, pd.DataFrame):
                Xo.iloc[test_ix,:] = \
                    self._fit_tes[ix].transform(X.iloc[test_ix,:])
            elif isinstance(X, np.ndarray):
                Xo[test_ix,:] = \
                    self._fit_tes[ix].transform(X[test_ix,:])
            else:
                raise TypeError('X must be DataFrame or ndarray')
        return Xo

            
    def fit_transform(self, X, y=None):
        """Fit and transform the data via target encoding.
        
        Parameters
        ----------
        X : pandas DataFrame, shape [n_samples, n_columns]
            DataFrame containing columns to encode
        y : pandas Series, shape = [n_samples]
            Target values (required!).

        Returns
        -------
        pandas DataFrame
            Input DataFrame with transformed columns
        """
        return self.fit(X, y).transform(X, y)
```

```python
# add tecv hyperparams on preproc pipeline
pipe_tecv = Pipeline(steps=[
    ('encoder', TargetEncoderCV()),
    ('scaler', MinMaxScaler()),
    ('imputer', SimpleImputer(strategy='mean'))])

preproc_tecv = ColumnTransformer(
    [('cat_preproc', pipe_tecv, cat_cols),
     ('num_preproc', SimpleImputer(strategy='constant', fill_value=0), num_cols)],
    remainder='passthrough')


# fit on train, transform train and test
preproc_tecv.fit(X_train, y_train)
X_train_tecv = preproc_tecv.transform(X_train)
X_test_tecv = preproc_tecv.transform(X_test)


#convert to dfs
X_train_tecv = pd.DataFrame(data=X_train_tecv, columns=X.columns)
X_test_tecv = pd.DataFrame(data=X_test_tecv, columns=X.columns)
```

## Define evaluation functions

Evaluate also twice:
- each model with each preproc, classification report
- each model with each preproc, main metrics and Confusion Matrix over test set

```python
def eval_report(models):

    for (n, model) in models:

        if 'ohe' in n.lower():
            model.fit(X_train_ohe, y_train)
            ytrain = y_train
            ytest = y_test
            ypred_train = model.predict(X_train_ohe)
            ypred_test = model.predict(X_test_ohe)

            
        if ' te ' in n.lower():
            model.fit(X_train_te, y_train)
            ytrain = y_train
            ytest = y_test
            ypred_train = model.predict(X_train_te)
            ypred_test = model.predict(X_test_te)
        
        
        if ' tecv' in n.lower():
            model.fit(X_train_tecv, y_train)
            ytrain = y_train
            ytest = y_test
            ypred_train = model.predict(X_train_tecv)
            ypred_test = model.predict(X_test_tecv)   
        
        print(n)
        print(f'Train Accuracy: {round(accuracy_score(ytrain, ypred_train)*100,3)} %')
        print((f'Test Accuracy: {round(accuracy_score(ytest, ypred_test)*100,3)} %'))
        print(f'Train F1: {round(f1_score(ytrain, ypred_train)*100,3)} %')
        print((f'Test F1: {round(f1_score(ytest, ypred_test)*100,3)} % \n'))
        print(n, 'classification report on test')
        print(classification_report(ytest, ypred_test), '\n\n')
```

```python
def eval_multi(models):

    seed = 42
    results = []
    names = []
    ypred_ohe=[]
    ypred_te=[]   
    
    scoring = {'accuracy' : make_scorer(accuracy_score),
               'f1_score' : make_scorer(f1_score),
               'roc_auc': make_scorer(roc_auc_score),
               'log_loss': make_scorer(log_loss)}

    for n, model in models:

        if 'ohe' in n.lower():      
            kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=seed)
            cv_r = cross_validate(model, X_train_ohe, y_train, cv=kfold, scoring=scoring)
            results.append(cv_r)
            names.append(n)
            ypred = model.fit(X_train_ohe, y_train).predict(X_test_ohe)
            ypred_ohe = ypred.copy()
            
        if ' te ' in n.lower():
            kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=seed)
            cv_r = cross_validate(model, X_train_te, y_train, cv=kfold, scoring=scoring)
            results.append(cv_r)
            names.append(n)
            ypred = model.fit(X_train_te, y_train).predict(X_test_te)
            ypred_te = ypred.copy()
            
        if ' tecv' in n.lower():
            kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=seed)
            cv_r = cross_validate(model, X_train_tecv, y_train, cv=kfold, scoring=scoring)
            results.append(cv_r)
            names.append(n)
            ypred = model.fit(X_train_tecv, y_train).predict(X_test_tecv)
            ypred_tecv = ypred.copy() 
            
    fig, ax = plt.subplots(4, 2, figsize=(12,13))
    fig.tight_layout(pad=4, rect=[0, 0.03, 1, 0.95])
    fig.suptitle('Models Comparison on cross validation', size=18)

    # build scores by metric
    acc = [results[i]['test_accuracy'] for i, r in enumerate(results)]
    f1 = [results[i]['test_f1_score'] for i, r in enumerate(results)]
    roc_auc = [results[i]['test_roc_auc'] for i, r in enumerate(results)]
    logloss = [results[i]['test_log_loss'] for i, r in enumerate(results)]
    
    # build cms
    cf_matrix_ohe = confusion_matrix(y_test, ypred_ohe)
    cf_matrix_te = confusion_matrix(y_test, ypred_te)
    cf_matrix_tecv = confusion_matrix(y_test, ypred_tecv)
    
    # plot
    sns.boxplot(data=acc, color='cornflowerblue', ax=ax[0,0]).set(xticklabels=names, title='accuracy on test')
    sns.boxplot(data=f1, color='cornflowerblue', ax=ax[0,1]).set(xticklabels=names, title='f1 on test')
    sns.boxplot(data=roc_auc, color='cornflowerblue', ax=ax[1,0]).set(xticklabels=names, title='roc_auc on test')
    sns.boxplot(data=logloss, color='cornflowerblue', ax=ax[1,1]).set(xticklabels=names, title='logloss on test')
    
    sns.heatmap(cf_matrix_ohe/np.sum(cf_matrix_ohe), annot=True, fmt='.2%', cmap='Blues', ax=ax[2,0]).\
    set_title('OHE CM on test')
    sns.heatmap(cf_matrix_te/np.sum(cf_matrix_te), annot=True, fmt='.2%', cmap='Blues', ax=ax[2,1]).\
    set_title('TE CM on test')
    sns.heatmap(cf_matrix_tecv/np.sum(cf_matrix_tecv), annot=True, fmt='.2%', cmap='Blues', ax=ax[3,0]).\
    set_title('TECV CM on test')
    ax[3,1].axis('off')
    
    return pd.DataFrame(results, names).apply(pd.Series.explode).reset_index().rename(columns={'index': 'model'})
```

## Neural network

```python
# plot function

def plot_metric_nn(history, metric):
    history_dict = history.history
    values = history_dict[metric]
    if 'val_' + metric in history_dict.keys():  
        val_values = history_dict['val_' + metric]

    epochs = range(1, len(values) + 1)

    if 'val_' + metric in history_dict.keys():  
        plt.plot(epochs, val_values, label='Validation')
    plt.semilogy(epochs, values, label='Training')

    if 'val_' + metric in history_dict.keys():  
        plt.title('Training and validation %s' % metric)
    else:
        plt.title('Training %s' % metric)
    plt.xlabel('Epochs')
    plt.ylabel(metric.capitalize())
    plt.legend()
    plt.grid()

    plt.show()  
```

```python
# prepare vectors

X_train_nn = X_train_tecv.values
y_train_nn = y_train.values
y_train_nn_tuple = to_categorical(y_train.values)

X_test_nn = X_test_tecv.values
y_test_nn = y_test.values
y_test_nn_tuple = to_categorical(y_test.values)
```

### 1st iteration

```python
# first approximation, 3 layers and softmax activation at last. Output is a tuple
# tanh activating first two layers shows the best performance

def build_model():
    m = models.Sequential()
    m.add(layers.Dense(32, input_shape=(X_train_nn.shape[1],), activation='tanh'))
    m.add(layers.Dense(16, activation='tanh'))
    m.add(layers.Dense(2, activation='softmax'))
    return m

# optimizer is rmsprop

model = build_model()

model.compile(
    optimizer = optimizers.RMSprop(),
    loss = losses.categorical_crossentropy,
    metrics = metrics.categorical_accuracy)

h = model.fit(X_train_nn,
              y_train_nn_tuple,
              epochs = 20,
              batch_size = 320,
              validation_split = .2)

plot_metric_nn(h, 'loss')
plot_metric_nn(h, 'categorical_accuracy')
```

```python
y_pred_nn = model.predict(X_test_nn)
y_pred_nn = y_pred_nn[:, 1]
y_pred_nn = (y_pred_nn > 0.5)
cm = confusion_matrix(y_test_nn, y_pred_nn)
print(cm)
```

```python
model.evaluate(X_test_nn, y_test_nn_tuple)
```

### 2nd iteration

```python
# from different optimizers on this neural network adamax optimizer works the best

model2 = build_model()
model2.compile(
    optimizer = 'adamax',
    loss = losses.categorical_crossentropy,
    metrics = metrics.categorical_accuracy)

h2 = model2.fit(X_train_nn,
              y_train_nn_tuple,
              epochs = 20,
              batch_size = 320,
              validation_split = .2)

plot_metric_nn(h2, 'loss')
plot_metric_nn(h2, 'categorical_accuracy')
```

```python
y_pred_nn2 = model2.predict(X_test_nn)
y_pred_nn2 = y_pred_nn2[:, 1]
y_pred_nn2 = (y_pred_nn2 > 0.5)
cm2 = confusion_matrix(y_test_nn, y_pred_nn2)
print(cm2)
```

```python
model2.evaluate(X_test_nn, y_test_nn_tuple)
```

### 3rd iteration

```python
# let's try a different network configuration, with sigmoid activation in the last layer, also adam optimizer

def build_model():
    m = models.Sequential()
    m.add(layers.Dense(32, input_shape=(X_train_nn.shape[1],), kernel_initializer='normal', activation='tanh'))
    m.add(layers.Dense(16, kernel_initializer='normal', activation='tanh'))
    m.add(layers.Dense(2, kernel_initializer='normal', activation='sigmoid'))
    return m

model3 = build_model()
model3.compile(
    optimizer = 'adam',
    loss = losses.categorical_crossentropy,
    metrics = metrics.categorical_accuracy)

h3 = model3.fit(X_train_nn,
              y_train_nn_tuple,
              epochs=20,
              batch_size=320,
              validation_split=.2)

plot_metric_nn(h3, 'loss')
plot_metric_nn(h3, 'categorical_accuracy')
```

```python
y_pred_nn3 = model3.predict(X_test_nn)
y_pred_nn3 = y_pred_nn3[:, 1]
y_pred_nn3 = (y_pred_nn3 > 0.5)
cm3 = confusion_matrix(y_test_nn, y_pred_nn3)
print(cm3)
```
